```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
zomato_df = pd.read_csv("C:/Users/ANJALI BABU/Downloads/zomato_restaurants.csv",sep=',', 
                     
                        header = 0)
zomato_df
```

```python
zomato_df.shape
```

```python
#A.Plot the Count of Famous Food Outlets / Restaurants in Bengaluru
plt.figure(figsize = (16,8))
famous=zomato_df['name'].value_counts()[:30]
sns.barplot(x=famous,y=famous.index,palette="dark")
plt.title("Most famous restaurants in Bangaluru",size = 20)
plt.xlabel("Number of outlets")
plt.ylabel('Name of Restaurants')
```

```python
#B.Display accepting online orders vs not accepting online orders count % distribution
plt.figure(figsize = (8,8))
plt.pie(zomato_df['online_order'].value_counts(), labels = zomato_df['online_order'].unique(), autopct='%1.2f%%', explode = [0,0])
plt.legend()
plt.show()
```

```python
#D.Display accepting Table Booking vs Not accepting table booking count % distribution
plt.figure(figsize = (8,8))
plt.pie(zomato_df['book_table'].value_counts(), labels = zomato_df['book_table'].unique(), autopct='%1.2f%%', explode = [0,0])
plt.legend()
plt.show()
```

```python
#C.Find the popular restaurant type in Bengaluru. 
plt.figure(figsize = (16,10))
restaurant = sns.countplot(zomato_df['rest_type'])
plt.xticks(rotation=90)
```

```python
#E.Histogram and Box Plot of Approx Cost Distribution.
#E.BOX PLOT
zomato_rest=zomato_df['approx_cost(for two people)'].str.replace(",", "")
plt.figure(figsize = (11,8))
sns.boxplot(data = zomato_rest,palette='plasma')
```

```python
#Histogram
zomato_df['approx_cost(for two people)'].hist(bins = 15, figsize=(35, 10),color ='darkorchid')
plt.title("Histogram plot of Approx_cost", frontsize =12)
plt.xlabel('approx_cost(for two people')
plt.ylabel('freequency (in range)')       
plt.show()                                             
```

```python
#g.Rate vs Approx Cost (for two people). Also categorise them into online_order available or not.
sns.set_style('whitegrid')
plt.figure(figsize=(20,17))
sns.scatterplot(x = "rate", y = "approx_cost(for two people)",data=zomato_df,palette='plasma',hue = 'online_order')
```

```python
# h.Distribution of Approx Cost for two people. Is it skewed?
cost= zomato_df['approx_cost(for two people)']
cost= cost.dropna().apply(lambda x : int(x.replace(',', '')))
plt.figure(figsize=(6,6))
sns.set_style('darkgrid')
sns.distplot(cost, color= 'green')
plt.title('Approx. cost for two people in city of Banguluru')
plt.xlabel('Cost for two people')
```

```python
#j.Which are the Top 20 most common restaurant types in Bengaluru?
plt.figure(figsize = (16,8))
rest_type=zomato_df['rest_type'].value_counts()[:20]
sns.barplot(x=rest_type,y=rest_type.index,palette="dark")
plt.title("Top 20 restaurants in Bangaluru",size = 20)
plt.xlabel("count")
plt.ylabel('Restaurants type')
```

```python
#k.Which are the Top 20 most common Locations with most Number of Restaurants?
plt.figure(figsize = (16,8))
location=zomato_df['location'].value_counts()[:20]
sns.barplot(x=location,y=location.index,palette="dark")
plt.title("Top 20 locations with most no.of restaurants in Bangaluru",size = 20)
plt.xlabel("count")
plt.ylabel('Restaurants location')
```

```python
#f.Histogram and Box Plot of Restaurant Rating Distribution. Note: Replace â€˜,â€™ from the values using replace(), lambda function with replace()
rating= zomato_df['rate']
rating= rating.dropna().apply(lambda x : float(x.split('/')[0]) if (len(x)>3) else np.nan).dropna()
plt.figure(figsize=(7,6))
sns.set_style('darkgrid')
sns.distplot(rating, bins = 20,  color= 'red')
plt.title('Rating distribution of the restaurants')
plt.xlabel('Ratings')
```
